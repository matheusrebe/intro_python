````python
#Exercício 3

import turtle # nos permite usar as tartarugas (turtles)
import math

jn = turtle.Screen()         # Configurar a janela e seus atributos
jn.bgcolor("lightgreen")

joana = turtle.Turtle() # Criar e configurar alguns atributos de joana
joana.hideturtle()
joana.color("red")
joana.pensize(3)
joana.speed(1000)

def arc(t, r, angle):
    from math import pi
    n = int(r)
    length = (((angle*pi)/180)*r)/n
    for i in range(n):
        t.forward(length)
        t.right(angle/n)  

def petala(t, length_petala, angle_petala):
    r = length_petala/(2*math.sin(math.radians(angle_petala)/2))
    for i in range(2):
      arc(t, r, angle_petala)
      t.right(180 - angle_petala)

def caule(t):
    t.color("darkgreen")
    t.right(90)
    t.forward(100)
    petala(t,length_petala = 35,angle_petala=70)
    t.forward(100)

def flower(t, length_petala, angle_petala, n_petalas):
    for i in range(n_petalas):
        petala(t, length_petala, angle_petala)
        t.right(360/n_petalas)
    caule(t)    
  
flower(joana, 70, 70, 7)
````

````python
#Exercício 5
import math
import turtle

jn = turtle.Screen()         # Configurar a janela e seus atributos
jn.bgcolor("lightgreen")

joana = turtle.Turtle() # Criar e configurar alguns atributos de joana
joana.hideturtle()
joana.color("red")
joana.pensize(1)
joana.speed(10)

def arquimedes(t, a, angle):
    t.penup()
    t.goto((a*angle)*math.cos(math.radians(angle)), (a*angle)*math.sin(math.radians(angle)))
    t.pendown()
    t.dot()

def arquimedes_spiral(t, a = 0.5, angle = 0, n =1):
    for angle in range(int(n)*420):
        arquimedes(t, a, angle)

arquimedes_spiral(joana)
````
